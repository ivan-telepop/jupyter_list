````python
 # Deijsktra decomposition 



# Начальные данные - Граф.
graph = {}
graph['start'] = {}
graph['a'] = {}
graph['b'] = {}
graph['fin'] = {}
# Назначаем веса узлов.
graph['start']['a'] = 6
graph['start']['b'] = 2
graph['a']['fin'] = 1
graph['b']['a'] = 3
graph['b']['fin'] = 5



# У ребер есть вес или цена
costs = {}
costs['a'] = 6
costs['b'] = 2
costs['fin'] = float('inf')



# Родители узлов/вершин
parents = {}
parents['a'] = "start"
parents['b'] = "start"
parents['in'] = None

# Пустой массив для обработаных узлов 
processed = list()


# достаем самый легкий вес
def find_low_cost(costs): # тут залетают цены (веса)
    lowest = float('inf') #дефолтные
    lowest_node = None #дефолтное
    for node in costs: # итеррируемся по списку весов/цен
        cost = costs[node] # проверяем его
        if cost < lowest and node not in processed: # если вес меньше чем... и его нет в результируещем
            lowest = cost
            lowest_node = node
    return lowest_node


node = find_low_cost(costs)
while node is not None:
    cost = costs[node]
    neighbors = graph[node]
    for nei in neighbors.keys():
        new_cost = cost + neighbors[nei]
        if costs[nei] > new_cost:
            costs[nei] = new_cost
            parents[nei] = node
    processed.append(node)
    node = find_low_cost(costs)


print(processed)
````

````python
Разбор полетов по алгоритму Дейкстры:


Допустим мы имеем ацикличный взвешенный граф:
Его вершины и веса:
ST >> A = 4
ST >> B = 1
ST >> F = 6
B >> F = 3
B >> C = 2
A >> C = 2
C >> F = 2
Так как пока мы не ослабили вершины, для нас все веса = inf (бесконечность)
Кроме ST который равен 0, потому что мы на старте и нам не надо никуда двигаться, а значит это равно 0
Нам необходима очередь куда мы будем записывать релакснутые (ослабленные) вершины: Обошли, записали в очередь !
** Ослабить вершину (релаксация) - переписать значение/вес с большего на меньшее.

Принцип обхода:
От старта к любой, из наших трех начальных: ST >> A , ST >> B , ST >> F
Первая итерация, кладем в очередь - узел, вес, и его предшественника.
Как выгдядит это:
    Релаксировали ST >> A = 4, ST >> B = 1, ST >> F = 6 
    Важно что в очереди на выход первым  - самый легкий вес, первый на извлечение и обход к следующим узлам.
    
    ------------------------------------
    | предшественник   | узел      | вес |
    --------------------------------------
    | ST               | B         | 1   |
    --------------------------------------
    | ST               | A         | 4   |
    --------------------------------------
    | ST               | F         | 6   |

Значит первым извлекаем ST >> B который весит 1, он у нас самый легкий и поэтому работаем с ним первым.

    B >> C = 2 + Вес ребра предшественника, он у нас = 1. То есть мы на прошлой иттерации определили вес ребра до узла В
    B >> F = 3 + Вес предшественника = 1
    И далее работаем по тому же алгоритму...
   ------------------------------------
    | предшественник   | узел      | вес |
    --------------------------------------
    | B                | C         | 3   |
    --------------------------------------
    | B                | F         | 4   |
    --------------------------------------
    Больше смежных ребер у B нет ! Следовательно переходим к другим, извлекаем из очереди самый легкий!
Дальше переходим к C >> F = 5, ребро до F не станем переписывать, оно тяжелее чем уже имеющееся значение ( B >> F  = 4 )
Возвращаемся к тем ребрам которые не иследованы.
A >> C  = 2 + (4 от ST >> A) = 6 - Мы не будем переписывать значение C = 6 потому что есть путь легче (S >> B >> C). 
Оставим C весом 3, а его предшественник B
C >> F = 2 + У узла C уже было значение 3 от ребер S >> B >> C, что в сумме дает вес 5, 
но мы знаем что до F мы доберемся за вес меньше 4 по ребрам от S >> B >> F
Так мы обошли все узлы, и нашли самый короткий путь от стартового - S, через - B, к - F финишу.






````

````python
# дейкстра ООП реализация

from __future__ import annotations


class Graph:
    def __init__(self, vertices: int) -> None:
        """
        >>> graph = Graph(2)
        >>> graph.vertices
        2
        >>> len(graph.graph)
        2
        >>> len(graph.graph[0])
        2
        """
        self.vertices = vertices
        self.graph = [[0] * vertices for _ in range(vertices)]

    def print_solution(self, distances_from_source: list[int]) -> None:
        """
        >>> Graph(0).print_solution([])  # doctest: +NORMALIZE_WHITESPACE
        Vertex 	 Distance from Source
        """
        print("Vertex \t Distance from Source")
        for vertex in range(self.vertices):
            print(vertex, "\t\t", distances_from_source[vertex])

    def minimum_distance(
        self, distances_from_source: list[int], visited: list[bool]
    ) -> int:
        """
        A utility function to find the vertex with minimum distance value, from the set
        of vertices not yet included in shortest path tree.

        >>> Graph(3).minimum_distance([1, 2, 3], [False, False, True])
        0
        """

        # Initialize minimum distance for next node
        minimum = 1e7
        min_index = 0

        # Search not nearest vertex not in the shortest path tree
        for vertex in range(self.vertices):
            if distances_from_source[vertex] < minimum and visited[vertex] is False:
                minimum = distances_from_source[vertex]
                min_index = vertex
        return min_index

    def dijkstra(self, source: int) -> None:
        """
        Function that implements Dijkstra's single source shortest path algorithm for a
        graph represented using adjacency matrix representation.

        >>> Graph(4).dijkstra(1)  # doctest: +NORMALIZE_WHITESPACE
        Vertex  Distance from Source
        0 		 10000000
        1 		 0
        2 		 10000000
        3 		 10000000
        """

        distances = [int(1e7)] * self.vertices  # distances from the source
        distances[source] = 0
        visited = [False] * self.vertices

        for _ in range(self.vertices):
            u = self.minimum_distance(distances, visited)
            visited[u] = True

            # Update dist value of the adjacent vertices
            # of the picked vertex only if the current
            # distance is greater than new distance and
            # the vertex in not in the shortest path tree
            for v in range(self.vertices):
                if (
                    self.graph[u][v] > 0
                    and visited[v] is False
                    and distances[v] > distances[u] + self.graph[u][v]
                ):
                    distances[v] = distances[u] + self.graph[u][v]

        self.print_solution(distances)


if __name__ == "__main__":
    graph = Graph(9)
    graph.graph = [
        [0, 4, 0, 0, 0, 0, 0, 8, 0],
        [4, 0, 8, 0, 0, 0, 0, 11, 0],
        [0, 8, 0, 7, 0, 4, 0, 0, 2],
        [0, 0, 7, 0, 9, 14, 0, 0, 0],
        [0, 0, 0, 9, 0, 10, 0, 0, 0],
        [0, 0, 4, 14, 10, 0, 2, 0, 0],
        [0, 0, 0, 0, 0, 2, 0, 1, 6],
        [8, 11, 0, 0, 0, 0, 1, 0, 7],
        [0, 0, 2, 0, 0, 0, 6, 7, 0],
    ]
    graph.dijkstra(0)

````
