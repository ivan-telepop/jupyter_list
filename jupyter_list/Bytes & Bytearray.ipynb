```python
# Maketrans table - INTRESTING THING !!!

# В общем прикольная штучня, когда нам нужно работать с заменой символов, можно создать две таблички
# Таблица А - содержит искомые символы (Те что ищем в байт строке)
# Таблица Б - символы на которые будут заменены символы из таблицы А

# Кротко,: все что нашли из таблицы А заменили на таблицу Б

# ВАЖНО Последовательность!!!


a = b'1234567890' # From sequnce - a

b = b'0987654321' # Symbols Will be changed to the symbls from sequnce - b

table = bytes.maketrans(a,b)

wrd = b'1234567890'

wrd.translate(table)
```
