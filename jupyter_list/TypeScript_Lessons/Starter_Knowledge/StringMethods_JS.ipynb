```python
1. String.prototype.charAt(index)
Описание: Возвращает символ по указанному индексу в строке.

Аргументы:
index (Number): Индекс символа (начиная с 0).
Возвращаемое значение: Символ на указанной позиции в строке. Если индекс выходит за пределы строки, возвращается пустая строка.
Пример:
let str = "Hello";
console.log(str.charAt(1)); // "e"



2. String.prototype.concat(...strings)
Описание: Объединяет строки.

Аргументы:
strings (String...): Строки, которые нужно объединить.
Возвращаемое значение: Новая строка, состоящая из исходной строки и всех переданных строк.
Пример:
let str1 = "Hello";
let str2 = "World";
console.log(str1.concat(" ", str2)); // "Hello World"


3. String.prototype.includes(searchString, position)
Описание: Проверяет, содержится ли подстрока в строке.
Аргументы:
searchString (String): Строка, которую нужно найти.
position (Number, optional): Индекс, с которого начинать поиск (по умолчанию 0).
Возвращаемое значение: true, если строка найдена, и false, если нет.
Пример:
let str = "Hello World";
console.log(str.includes("World")); // true
console.log(str.includes("world")); // false
4. String.prototype.indexOf(searchValue, start)
Описание: Возвращает индекс первого вхождения подстроки.

Аргументы:
searchValue (String): Подстрока для поиска.
start (Number, optional): Индекс, с которого начинать поиск (по умолчанию 0).
Возвращаемое значение: Индекс первого вхождения подстроки или -1, если подстрока не найдена.
Пример:
let str = "Hello World";
console.log(str.indexOf("o")); // 4
console.log(str.indexOf("o", 5)); // 7
5. String.prototype.slice(beginIndex, endIndex)

Аргументы:
beginIndex (Number): Индекс начала извлечения (включительно).
endIndex (Number, optional): Индекс конца извлечения (не включительно).
Возвращаемое значение: Новая строка, содержащая извлечённую подстроку.
Пример:
let str = "Hello World";
console.log(str.slice(0, 5)); // "Hello"
console.log(str.slice(6)); // "World"
6. String.prototype.split(separator, limit)
Описание: Разделяет строку на массив строк.

Аргументы:
separator (String | RegExp): Разделитель, по которому строка будет разделена.
limit (Number, optional): Максимальное количество элементов в результирующем массиве.
Возвращаемое значение: Массив строк.
Пример:
let str = "apple,banana,orange";
console.log(str.split(",")); // ["apple", "banana", "orange"]
console.log(str.split(",", 2)); // ["apple", "banana"]
7. String.prototype.replace(searchValue, newValue)
Описание: Заменяет первое вхождение подстроки.

Аргументы:
searchValue (String | RegExp): Строка или регулярное выражение, которое нужно заменить.
newValue (String): Новое значение для замены.
Возвращаемое значение: Новая строка с заменённым значением.
Пример:

javascript
Copy
Edit
let str = "Hello World";
console.log(str.replace("World", "JavaScript")); // "Hello JavaScript"
8. String.prototype.replaceAll(searchValue, newValue)
Описание: Заменяет все вхождения подстроки.

Аргументы:
searchValue (String | RegExp): Строка или регулярное выражение, которое нужно заменить.
newValue (String): Новое значение для замены.
Возвращаемое значение: Новая строка с заменёнными всеми значениями.
Пример:

javascript
Copy
Edit
let str = "Hello World, Hello!";
console.log(str.replaceAll("Hello", "Hi")); // "Hi World, Hi!"
9. String.prototype.trim()
Описание: Удаляет пробелы с начала и конца строки.

Аргументы: Нет.
Возвращаемое значение: Новая строка без пробелов по бокам.
Пример:

javascript
Copy
Edit
let str = "   Hello World   ";
console.log(str.trim()); // "Hello World"
10. String.prototype.toLowerCase()
Описание: Преобразует строку в нижний регистр.

Аргументы: Нет.
Возвращаемое значение: Новая строка в нижнем регистре.
Пример:

javascript
Copy
Edit
let str = "Hello World";
console.log(str.toLowerCase()); // "hello world"
11. String.prototype.toUpperCase()
Описание: Преобразует строку в верхний регистр.

Аргументы: Нет.
Возвращаемое значение: Новая строка в верхнем регистре.
Пример:

javascript
Copy
Edit
let str = "Hello World";
console.log(str.toUpperCase()); // "HELLO WORLD"
12. String.prototype.startsWith(searchString, position)
Описание: Проверяет, начинается ли строка с указанной подстроки.

Аргументы:
searchString (String): Подстрока для проверки.
position (Number, optional): Индекс, с которого начинать проверку (по умолчанию 0).
Возвращаемое значение: true, если строка начинается с указанной подстроки, иначе false.
Пример:

javascript
Copy
Edit
let str = "Hello World";
console.log(str.startsWith("Hello")); // true
console.log(str.startsWith("World", 6)); // true
13. String.prototype.endsWith(searchString, length)
Описание: Проверяет, заканчивается ли строка на указанную подстроку.

Аргументы:
searchString (String): Подстрока для проверки.
length (Number, optional): Длина строки, которая будет проверяться (по умолчанию равна длине строки).
Возвращаемое значение: true, если строка заканчивается на указанную подстроку, иначе false.
Пример:

javascript
Copy
Edit
let str = "Hello World";
console.log(str.endsWith("World")); // true
console.log(str.endsWith("Hello", 5)); // true
14. String.prototype.match(regexp)
Описание: Находит все совпадения с регулярным выражением.

Аргументы:
regexp (RegExp): Регулярное выражение для поиска.
Возвращаемое значение: Массив всех совпадений или null, если совпадений не найдено.
Пример:

javascript
Copy
Edit
let str = "The rain in Spain stays mainly in the plain";
let matches = str.match(/ain/g);
console.log(matches); // ["rain", "Spain", "mainly", "plain"]
```
