```python
# wraps() - Это функция которая применяется к функции обертке для подстановки атрибутов оборачиваемой функции
# @wraps(wrapped_func, assigned
from functools import wraps

def my_decorator(func):
    def wrapper(*args,**kwargs):
        """w r a p p e r"""
        func()
    
    return wrapper    


@my_decorator
def main(name):
    """m a i n  -   f u n c"""
    return {"name":name}

@my_decorator
def hello(dev):
    """ H E L L O   func"""
    return dev * 4



print(hello.__doc__,hello.__name__) 
print(main.__name__,main.__doc__)
# The problem is that when we call __doc__,__name__ of decorated func we obtaining wrapper functions __doc__
# Theoretically we me substitute inside wrapper these params but it not convenient way to sovle this issue
def second_decorator(func):
    def wrapper(*args, **kwargs):
        """A D E C O R A T O R """
        func()
    wrapper.__name__ = func.__name__
    wrapper.__doc__ = func.__doc__
    return wrapper

@second_decorator
def main(name):
    """m a i n  -   f u n c"""
    return {"name":name}

@second_decorator
def hello(dev):
    """ H E L L O   func"""
    return dev * 4
print(hello.__doc__,hello.__name__) 
print(main.__name__,main.__doc__)

# what if we will do type help ?
print("\n\n")
help(hello)

# Its better to save time and nerv system we will use wraps decor - lets look on it
def ok_decorator(func):
    @wraps(func)
    def wrapper(*args,**kwargs):
        """w r a p p e r"""
        func()
    
    return wrapper 


@ok_decorator
def good(good):
    """-=here is so good=-  """
    return 100


@ok_decorator
def love(love):
    """-=LOVE=-  """
    return "hearts"


print("\n\n\n")

help(good)
print("\n")

print(love.__doc__)
```
