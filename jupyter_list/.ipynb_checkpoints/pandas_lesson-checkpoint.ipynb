#  Copy | Concat | 

```python
array = {'name':'DF_1','a':[1,2,3],'b':123,'c':0,'d':888,'e':'fff'}
ar_two = {'name':'DF_2','a':[5555],'b':8888,'c':0,'d':2222,'e':'fff'}
# TWO ARRAYS up there

df_one = pd.DataFrame.from_dict(array)
df_two = pd.DataFrame.from_dict(ar_two)
# Created DataFrame objects

df_2 = df_one.copy(deep=True) # deep=true - that means that is not a linked to original obj, it's absolutely another object

# That's how we concat these
both_obj = pd.concat([df_one,df_two])
both_obj.shape
```
