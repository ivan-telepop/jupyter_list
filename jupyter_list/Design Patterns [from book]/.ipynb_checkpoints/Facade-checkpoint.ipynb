```python
from enum import Enum
from abc import ABCMeta,abstractmethod


State = Enum('State','new running envoked')

class Server(metaclass=ABCMeta):
    @abstractmethod
    def __init__(self):
        pass
    def __str__(self):
        return self.name
    @abstractmethod
    def boot(self):
        pass
    @abstractmethod
    def kill(self,restart=True):
        pass



class FileServer(Server):
    def __init__(self):
        self.name = 'FileServer'
        self.state = State.new
    def boot(self):
        print('Server is running')
        self.state = State.running
    def kill(self, restart=True):
        print('killing server')
        self.state = State.restart if restart else State.waiting
    def create_file(self,user,name,premiss):
        print(f"Trying create file {user} {name} {premiss}")


class ProcessServer(Server):
    def __init__(self):
        self.name = 'Process Server'
        self.state = State.new

    def boot(self):
        print('Booting process server')
        self.state = State.running
    def kill(self,restart=True):
        print('Killing kernel process')
        self.state = State.restart if restart else State.waiting
    def create_process(self,user,name):
        print(f'Trying runn the process :{user}  by {name}')

class OperatingSystem:
 '''The Facade'''
     def __init__(self):
         self.fs = FileServer()
         self.ps = ProcessServer()
     def start(self):
         [i.boot() for i in (self.fs, self.ps)]
     def create_file(self, user, name, permiss):
         return self.fs.create_file(user, name, permiss)
     def create_process(self, user, name):
         return self.ps.create_process(user, name)
```
